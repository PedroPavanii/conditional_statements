## **1) Quais as espécies que mais atacaram no Brasil nos ultimos 10 anos?**

```python
#ESTABELECENDO FILTROS PARA RESPONDER A QUESTÃO ACIMA.
especies_brasil_a = tabela_shark.loc[:,['Country','Year', 'Species ']]
especies_brasil_b = especies_brasil_a.loc[(especies_brasil_a['Country'] == 'BRAZIL') & (especies_brasil_a['Year'] >= 2003.)] 
especies_c =especies_brasil_b.dropna()


#RETIRANDO TODOS OS NULOS, NOMES DE ESPECIES ERRADOS E VALORES NÃO PERTENCENTES.
especies_d = especies_c[especies_c['Species '].isin(['Lemon shark pup','dogfish','Lemon shark', 'Tiger shark', 'small shark', 'Sandtiger shark','Tiger shark', 'Bull shark',' tiger shark', 'bull shark'])]
especies_final = especies_d.reset_index(drop = True)


#MELHORANDO ESTETICA DA APRESENTAÇÃO DOS VALORES NAS COLUNAS.
especies_final['Species '] = especies_final['Species '].str.title()


#TRADUZINDO AS COLUNAS PARA PORTUGUÊS.
especies_final = especies_final.rename(columns={'Country': 'País', 'Species ': 'Espécies', 'Year': 'Ano'})


#RETIRANDO A COLUNA PAÍS.
especies_final = especies_final.drop('País', axis = 1)
especies_final = especies_final.drop('Ano', axis = 1)


#CRIANDO UM DATAFRAME PARA RELACIONAR ESPECIES POR ANO.
df_especies_final=especies_final.value_counts().reset_index()
df_especies_final.columns = [ 'Espécies', 'Nº de Ataques']


df_especies_final
#FALTA GRAFICO
```

## **2) Quais as praias do brasil que mais tiveram incidentes?**

```python
#FILTRANDO AS PRAIAS DO BRASIL POR AREA E LOCALIZAÇÃO. RETIRANDO VALORES NULOS E LINHAS COM RESULTADOS NÃO PERTENCENTES.
praias = tabela_shark.loc[:,['Country','Area','Location']]
praias_br = praias.loc[praias['Country'] =='BRAZIL'].dropna().reset_index(drop=True)
praias_br.drop(index =[110, 13, 20, 36], inplace = True)


#REALIZANDO FILTRAGEM, SUBSTITUIÇÃO DE VALORES VISANDO MELHORA ESTETICA E MELHOR VISUALIZAÇÃO.
praias_br['Area']= praias_br['Area'].str.replace(r'\.', ' ', regex=True)
praias_br['Location']= praias_br['Location'].str.replace(r',.*', '', regex=True)
praias_br['Location']= praias_br['Location'].str.replace('Punta Del Chifre', 'Praia Del Chifre')
praias_br['Location']= praias_br['Location'].str.replace('Beach', '')
praias_br['Location']= praias_br['Location'].str.replace('beach', '')
praias_br = praias_br.drop('Country', axis=1)


#CRIANDO UM DATAFRAME QUE RELACIONE AS PRAIAS COM OS INCIDENTES.
df_praia = praias_br['Location'].value_counts().reset_index()
df_praia.columns = ['Praia', 'Nº de Ataques']


df_praia


#praias repetidas nao agruparam
#FALTA GRAFICO

```
