```python
#Especies que mais atacaram no Brasil nos ultimos 10 anos

import matplotlib.pyplot as plt
especies_brasil_a = tabela_shark.loc[:,['Country','Species ','Year']]

especies_brasil_b = especies_brasil_a.loc[(especies_brasil_a['Country'] == 'BRAZIL') & (especies_brasil_a['Year'] >= 2003.)] 

especies_c =especies_brasil_b.dropna()

especies_d = especies_c[especies_c['Species '].isin(['Lemon shark pup','dogfish','Lemon shark', 'Tiger shark', 'small shark', 'Sandtiger shark','Tiger shark', 'Bull shark',' tiger shark', 'bull shark'])]
especies_e = especies_d.reset_index(drop = True)

especies_e['Species '] = especies_e['Species '].str.title()

especies_e.rename(columns={'Country': 'Pa√≠s', 'Species ': 'Esp√©cies', 'Year': 'Ano'})

#falta grafico
```
